```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency

hppd=pd.read_csv('train.csv')
hppd.head()
```

```python
hppd['SalePrice'].describe()
```

```python
sns.histplot(hppd['SalePrice'], color='blue', kde=True)
plt.title('Distribution of Sale Price')
plt.show()
```

```python
correlationm = hppd.corr(numeric_only=True)
correlationm['SalePrice'].sort_values(ascending=False)
```

```python
plt.figure(figsize=(20, 10))
sns.heatmap(correlationm, cmap='coolwarm', annot=True, fmt='.2f')
plt.title("Correlation Heatmap")
plt.show()
```

```python
k=10 #Top 10 highly correlated Features
cols=correlationm.nlargest(k,'SalePrice')['SalePrice'].index
cm=np.corrcoef(hppd[cols].values.T)
plt.figure(figsize=(10, 6))
sns.heatmap(cm, cmap='coolwarm', annot=True, fmt='.2f',xticklabels=cols.values, yticklabels=cols.values)
plt.title("Correlation Heatmap")
plt.show() #Gets to know about Multicollinearity. Shows GarageCars and GarageArea are also correlated to each other so we can choose one.
```

#### 1) Determine how Overall Material and Finish Quality affects the Property's Sale Price.

```python
plt.figure(figsize=(10, 6))
sns.histplot(hppd['OverallQual'], kde=True, color='red')
plt.title('Distribution of Overall Material and Finish Quality')
plt.xlabel('Overall Quality')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 7))
sns.regplot(x=hppd['OverallQual'],y=hppd['SalePrice'], fit_reg=True, color='blue')
plt.title('Relationship between Sale Price and Overall Quality')
plt.xlabel('Overall Quality')
plt.ylabel('Sale Price')
plt.show()
```

```python
corr1 = hppd['SalePrice'].corr(hppd['OverallQual'])
f'The Correlation between SalePrice and OverallQual: {corr1:.2f}'
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=hppd['OverallQual'], y=hppd['SalePrice'], palette='Set1')
plt.title('Sale Price vs. Overall Quality')
plt.xlabel('Overall Quality')
plt.ylabel('Sale Price')
plt.show()
```

```python
overallqualprice = hppd.groupby('OverallQual')['SalePrice'].mean()
plt.figure(figsize=(10, 6))
sns.barplot(x=overallqualprice.index, y=overallqualprice.values, palette='Set1')
plt.title('Mean Sale Price for Each Overall Quality Category')
plt.xlabel('Overall Quality')
plt.ylabel('Mean Sale Price')
plt.show()
```

#### 3) Determine if Original Construction Date affects the Property's Sale Price.

```python
plt.figure(figsize=(10, 6))
sns.histplot(hppd['YearBuilt'], kde=True, color='Red')
plt.title('Distribution of Original Construction Year')
plt.xlabel('Year Built')
plt.show()
```

```python
plt.figure(figsize=(10, 7))
sns.regplot(x=hppd['YearBuilt'],y=hppd['SalePrice'], fit_reg=True, color='Blue')
plt.title('Relationship between Sale Price and Original Construction Date')
plt.xlabel('Year Built')
plt.ylabel('Sale Price')
plt.show()
```

#### 4) Determine how the Number of Sales have changed over the Years

```python
yearly_count = hppd['YrSold'].value_counts().sort_index()
plt.subplot(1,2,1)
sns.barplot(x=yearly_count.index, y=yearly_count.values, palette="Set1")
plt.title('Bar Graph of Number of Sales in Each Year')
plt.xlabel('Year Built')
plt.ylabel('Number of Sales')
plt.xticks(rotation=90)

plt.subplot(1,2,2)
plt.pie(yearly_count, labels=yearly_count.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('Blues'))
plt.title('Proportion of Sales by Year')

plt.suptitle("Number of Sales in Each Year")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
yearly_count.plot(color='Red')
plt.title('Number of Sales Over the Years')
plt.xlabel('Year Built')
plt.ylabel('Number of Sales')
plt.show()
```

**Relationship between Sale Price by LotShape and LandContour**

```python
plt.figure(figsize=(12, 6))
ax = sns.barplot(data=hppd, x='LotShape', y='SalePrice', hue='LandContour', palette='Set1')
plt.title('Sale Price by LotShape and LandContour')
plt.xlabel('LotShape')
plt.ylabel('Sale Price')
plt.legend(title='LandContour')
plt.xticks(rotation=0)
plt.show()
```

```python
bin_edges = [0, 100000, 200000, 800000]
bin_labels = ['Low', 'Medium', 'High']
hppd['SalePriceCategory'] = pd.cut(hppd['SalePrice'], bins=bin_edges, labels=bin_labels)
contingency_table = pd.crosstab(index=hppd['LotShape'], columns=hppd['LandContour'], values=hppd['SalePriceCategory'], aggfunc='count')

chi2, p, _, _ = chi2_contingency(contingency_table)
print(f'Chi-squared value: {chi2:.2f}')
print(f'p-value: {p:.15f}')
```
